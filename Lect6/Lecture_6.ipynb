###Parsing gff file and put into 

```python
import sys, glob
Datastore = []
for filename in glob.glob('*.gff'):
        f = open(filename, 'r')
        content = f.readlines()
        f.close
        for line in content:
                if (line[0]!='#'):
                        seperate = line.split()
                        filename= seperate[0]
                        classification = seperate[2]
                        start = seperate[3]
                        end = seperate[4]
                        id = seperate[8]    
                        items = id.split(';')
                        keys = {}
                        for item in items:
                            temp = item.split('=')
                            keys[temp[0]]=temp[1]
                            
                        if 'product' in keys:
                            product = keys['product']
                        else:
                            product = ''
                        if 'gene' in keys:
                            gene = keys['gene']
                        else:
                            gene = ''
                        if 'Name' in keys:
                            name = keys['Name']
                        else:
                            name = ''
                        if 'Note' in keys:
                            note = keys['Note']
                        else:
                            note=''                          
                        record ={
                                'filename':filename,
                                'classification':classification,
                                'start':start,
                                'end':end,
                                'gene':gene,
                                'name':name,
                                'note':note
                                 }
                        Datastore.append(record)
```

```python
import pandas as pd
import numpy as np
plotdata = pd.read_csv('Data.csv',header=0)
plotdata.columns = ['A','B','C','D','E','F','G','H','I','J','K']
plotdata
```

```python
import numpy as np
time = np.arange(0, float(plotdata['A'].count()), 1)
timeS = pd.Series(time)
timeS
datawithtime = pd.concat([timeS, plotdata], axis=1)
datawithtime.columns = ['Time','A','B','C','D','E','F','G','H','I','J','K']
datawithtime
```

```python
import math
average = plotdata.mean(axis=1)
stdev = plotdata.std(axis=1)
stdev
stderr = stdev / math.sqrt(11)
stderr
data_descript = pd.concat([datawithtime,average,stderr],axis=1) 
data_descript.columns = ['time','A','B','C','D','E','F','G','H','I','J', 'K','mean','stderr']
plot = data_descript.plot(x='time',y='mean',yerr='stderr')
fig = plot.get_figure()
fig.savefig('myplot.pdf')
```
